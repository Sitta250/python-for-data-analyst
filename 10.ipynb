```python
import pandas as pd

# Creating a sample dataset
data = {
    'job_title': ['Software Engineer', 'Senior Data Scientist', 'Junior Developer', 
                  'Lead Engineer', 'Intern', 'Data Analyst', 'Software Engineer'],
    'salary_year_avg': [85000, 120000, 55000, 130000, 30000, 70000, 110000]
}

df_level_salary = pd.DataFrame(data)
print(df_level_salary)
```

```python
def exp_cat(row):
  if 'Senior' in row['job_title'] or 'Lead' in row['job_title']:
    return 'Senior'
  elif 'Junior' in row['job_title'] or 'Inter' in row['job_title']:
    return 'Entry'
  elif row['salary_year_avg']>100_000:
    return 'Senior'
  elif row['salary_year_avg']<50_000:
    return 'Entry'
  else:
    return 'Mid'

df_level_salary['experience level'] = df_level_salary.apply(exp_cat, axis=1)
df_level_salary
```
