```python
import pandas as pd
from datasets import load_dataset
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.sample(5)
```

```python
df.index
```

```python
df.index.name = 'job_index'
```

```python
df.head()
```

```python
df.index.dtype
```

```python
df_usa = df[df['job_country']=='United States']
df_usa.reset_index(inplace = True)
df_usa
```

```python
df_usa.set_index('job_index', inplace = True)
df_usa
```

```python
median_pivot = df_usa.pivot_table(
  values='salary_year_avg', 
  index='job_title_short',
  aggfunc=['median', 'min', 'max']
)
median_pivot
```

```python
median_pivot.sort_values(by=('median', 'salary_year_avg'), ascending=False)
```

```python
#sorting data by job_title_short
median_pivot.sort_index(inplace=True)
median_pivot
```

```python
df_germany = df[df['job_country']=='Germany']
df_germany
```

```python
df_germany.reset_index(inplace=True)
df_germany
```

```python
Germany_pivot_table = df_germany.pivot_table(
  values='salary_year_avg', 
  index='job_title_short',
  aggfunc=['median', 'mean', 'min', 'max']
)
Germany_pivot_table
```

```python
Germany_pivot_table.sort_index(inplace=True)
Germany_pivot_table
```

```python
Germany_pivot_table.sort_values(by=('median', 'salary_year_avg'), ascending=False)
```
